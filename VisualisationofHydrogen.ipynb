```python
import plotly.express as px
import pandas as pd

# Extract summary statistics for the 'H_ppm' column
stats_series = df['H_ppm'].describe()

# Create a DataFrame for visualization
stats_df = pd.DataFrame({
    'Statistic': ['mean', 'std', 'min', '25%', '50%', '75%', 'max'],
    'Value': [stats_series['mean'], stats_series['std'], stats_series['min'],
              stats_series['25%'], stats_series['50%'], stats_series['75%'], stats_series['max']]
})

# Bar plot with a custom color scale
fig = px.bar(stats_df, x='Statistic', y='Value',
             title='Summary Statistics of H_ppm',
             labels={'Statistic': 'Statistic', 'Value': 'Value'},
             color='Value',
             color_continuous_scale='Viridis',  # Choose a color scale of your preference
             height=400)

# Customize layout
fig.update_layout(xaxis_title='Statistic for the H_ppm', yaxis_title='Value', coloraxis_showscale=False)
fig.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming 'spectral_features' is your dataset
# Choose the number of clusters (k)
k = 3

# Create a KMeans instance
kmeans = KMeans(n_clusters=k, random_state=42)

# Extracting Specific Features
X = spectral_features.iloc[:, :-1]  # Exclude the 'Cluster' column

# Convert column names to strings
X.columns = X.columns.astype(str)

# Fit the model to your data
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Add cluster labels to your DataFrame
spectral_features['Cluster'] = labels
```

- Explain the purpose of the clustering analysis and why it's important.
Highlight any patterns or insights gained from the clustering results.
Discuss the implications of the concentration being higher in the center.

- This 3D scatter plot illustrates the results of K-means clustering on our spectral features dataset. Each point represents a data point, with colors indicating its cluster assignment. The higher concentration of points in the center suggests a potential central tendency or grouping of similar samples. The clustering analysis provides insights into natural groupings within the dataset, shedding light on patterns and relationships that may have significant implications for our understanding of spectral features.

```python
fig = go.Figure()

scatter = go.Scatter3d(
    x=X.iloc[:, 0],
    y=X.iloc[:, 1],
    z=X.iloc[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color=labels,
        colorscale='Viridis',
        opacity=0.8,
        line=dict(color='black', width=0.5)
    )
)

fig.add_trace(scatter)

fig.update_layout(
    scene=dict(
        xaxis_title='Feature 1',
        yaxis_title='Feature 2',
        zaxis_title='Feature 3'
    ),
    title='K-Means Clustering',
    margin=dict(l=0, r=0, b=0, t=30),
    showlegend=False
)

# Adding a color bar legend
colorbar = go.ColorBar(
    title='Cluster',
    tickvals=list(set(labels)),  # Assuming labels are integers representing clusters
    ticktext=[str(cluster) for cluster in set(labels)],
    ticks='outside'
)

fig.update_layout(coloraxis_colorbar=colorbar)

fig.show()
```

Slide Content:

- Objective:

     Understand the importance of features in predicting 'H_ppm' and 'Meth_ppm' levels using a RandomForestRegressor.

- Data Preparation:

     Extract features from 'spectral_features' excluding 'Cluster', 'H_ppm', and 'Meth_ppm'.
     Prepare labels by extracting 'H_ppm' and 'Meth_ppm' columns.

- Model Training:

     Train a RandomForestRegressor with 5 trees.

- Feature Importance Analysis:

   Assess the importance of features in predicting 'H_ppm' and 'Meth_ppm'.

- Results:

  Display feature labels and sorted indices based on their importance.

```python
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'spectral_features' is your dataset
# Exclude the 'Cluster' column for training
X = spectral_features  # Exclude 'Cluster', 'H_ppm', 'Meth_ppm' columns

# Extract the 'H_ppm' and 'Meth_ppm' columns for labeling
labels = df[['H_ppm', 'Meth_ppm']]

# Convert column names to strings
X.columns = X.columns.astype(str)

# Create a Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=5, random_state=42)

# Fit the model to your data
rf_regressor.fit(X, labels)

# Get feature importances from the trained model
tree_feature_importances = rf_regressor.feature_importances_

# Get feature labels
feature_labels = X.columns

# Sort features by importance
sorted_idx = tree_feature_importances.argsort()

```

```python
# Set the number of top features to display
top_features = 20

# Create subplots
fig = go.Figure()

# Plotting for 'H_ppm'
fig.add_trace(go.Bar(
    y=feature_labels[sorted_idx][-top_features:],
    x=tree_feature_importances[sorted_idx][-top_features:],
    orientation='h',
    name="Top Features for 'H_ppm'",
    marker=dict(color='skyblue')
))

# Plotting for 'Meth_ppm'
fig.add_trace(go.Bar(
    y=feature_labels[sorted_idx][-top_features:],
    x=tree_feature_importances[sorted_idx][-top_features:],
    orientation='h',
    name="Top Features for 'Meth_ppm'",
    marker=dict(color='lightgreen')
))

# Customize layout
fig.update_layout(
    title="Top Features Influencing 'H_ppm' and 'Meth_ppm'",
    xaxis=dict(title="Feature Importance"),
    yaxis=dict(title="Feature Labels"),
    barmode='stack',
    height=600,
    width=1000
)

fig.show()
```

Key Points for Presentation:

- Objective:

 Compare the feature importance of top features for 'H_ppm' and 'Meth_ppm'.

- Data Preparation:

 Identify the top features based on their importance values.
- Visualization:

 A horizontal stacked bar chart is used to represent the feature importance of the top features for both 'H_ppm' and 'Meth_ppm'.
 Colors (skyblue and lightgreen) distinguish between the two variables.

- Insights:

 Discuss any noticeable differences in feature importance between 'H_ppm' and 'Meth_ppm'.
 Use this visualization to highlight key predictors for each variable.

- Implications:

 Explain how understanding feature importance can guide further analysis or decision-making in the context of 'H_ppm' and 'Meth_ppm'.

```python
# Choose the number of top features to include in the 3D plot
num_top_features = 3

# Select the top features for 3D plotting
top_features_indices = sorted_idx[-num_top_features:]

# Create a 3D scatter plot for the selected top features
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot using the selected top features
scatter = ax.scatter(
    X.iloc[:, top_features_indices[0]],
    X.iloc[:, top_features_indices[1]],
    X.iloc[:, top_features_indices[2]],
    c=tree_feature_importances[top_features_indices],
    cmap='viridis',
    marker='o'
)

# Adding labels and title
ax.set_xlabel(feature_labels[top_features_indices[0]])
ax.set_ylabel(feature_labels[top_features_indices[1]])
ax.set_zlabel(feature_labels[top_features_indices[2]])
ax.set_title(f"Top {num_top_features} Features in 3D")

# Adding color bar
cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Feature Importance')

# Show the plot
plt.show()
```
